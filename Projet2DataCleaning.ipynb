```python
#Some columns are sill useless, we only need numerical values

base_num = new_database._get_numeric_data()  
base_num.head()                             
```

```python
#Creating base_num_reduite that does not contain numerical but useless columns : code, created_t et last_modified_t

base_num_reduite = base_num.drop(['code','created_t','last_modified_t'], axis = 1) 
```

```python
# We want to replace NaN values by the mean of this column values, this way, if the column contains lots of zeros,
# The value we add will be close to zero and would not impact too much in the case of statistical analyses
# If the column contains few zeros, then adding values corresponding to the column mean will have less impact than adding a zero

moyenne = []
for i in base_num_reduite:
    temp = []
    compteur = 0
    for j in base_num_reduite[i]:
        if np.isnan(j) == False:   
            temp.append(j)
    moyenne_temp = np.mean(np.array(temp))
    print(moyenne_temp)

    for k in base_num_reduite[i]:    
        if np.isnan(k) == True:       
            base_num_reduite[i].iloc[compteur] = moyenne_temp
            compteur = compteur +1
        else:
            compteur = compteur +1 
  
```
